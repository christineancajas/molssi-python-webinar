```python
# generate general pathname for the .out file

import os

homework_file = os.path.join('data','03_Prod.mdout')
print(homework_file)
```

```python
# open and read the .out file, then close
outfile = open(homework_file,'r') 
data = outfile.readlines() 
outfile.close()

#loop test to see if I can extract the energy associated with Etot for each step
for line in data:
    
    if 'NSTEP' in line:
        step_line = line
        step_split = step_line.split()
        step = step_split[2]
        
    if 'Etot' in line:
        energy_line = line
        etot_words = energy_line.split()
        energy = float(etot_words[2])    
        
        print(F'{step}\t:\t{energy}\n')
        
```

```python
import os

homework_file = os.path.join('data','03_Prod.mdout')

# make or open file Etot.txt in which the data extracted from .out will be saved
filehandle = open('Etot.txt','w+')

# open and read the .out file, then close
outfile = open(homework_file,'r') 
data = outfile.readlines() 
outfile.close()                   

for line in data:
    if 'NSTEP' in line:
        step_line = line
        step_split = step_line.split()
        step = step_split[2]
    
    if 'Etot' in line:
        energy_line = line
        etot_words = energy_line.split()
        energy = float(etot_words[2])
        
        
        filehandle.write(F' {step}\t {energy} \n')
                         
filehandle.close()
```
